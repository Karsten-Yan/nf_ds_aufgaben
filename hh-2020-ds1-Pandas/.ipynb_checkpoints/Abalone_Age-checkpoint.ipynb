```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

df = pd.read_csv("abalone.data", delimiter=",", header=None)
cols = ["Sex","Length","Diameter","Height","Whole_weight","Shucked_weight","Viscera_weight","Shell_weight","Rings"]
df.columns = cols
df.head()
```

```python
df.eval("Age = Rings + 1.5", inplace = True)
df.head()
```

```python
plt.scatter(df.Age, df.Diameter)
plt.title("Age vs Diameter")
plt.xlabel("Age [years]")
plt.ylabel("Diameter [mm]")
```

```python
plt.scatter(df.query("Sex == 'M'").Age, df.query("Sex == 'M'").Diameter)
plt.title("Age vs Diameter Male Abalone")
plt.xlabel("Age [years]")
plt.ylabel("Diameter [mm]")
```

```python
plt.scatter(df.query("Sex == 'F'").Age, df.query("Sex == 'F'").Diameter)
plt.title("Age vs Diameter Female Abalone")
plt.xlabel("Age [years]")
plt.ylabel("Diameter [mm]")
```

```python
plt.scatter(df.query("Sex == 'I'").Age, df.query("Sex == 'I'").Diameter)
plt.title("Age vs Diameter Infant Abalone")
plt.xlabel("Age [years]")
plt.ylabel("Diameter [mm]")
```

```python
plt.scatter(df.Age, df.Length)
plt.title("Age vs Length")
plt.xlabel("Age [years]")
plt.ylabel("Length [mm]")
```

```python
plt.scatter(df.Diameter, df.Length)
plt.title("Diameter vs Length")
plt.xlabel("Diameter [mm]")
plt.ylabel("Length [mm]")
```

```python
plt.scatter(df.Diameter, df.Height)
plt.title("Diameter vs Height")
plt.xlabel("Diameter [mm]")
plt.ylabel("Height [mm]")
```

```python
df_sex_group = df.groupby("Sex")
df_sex_group.Whole_weight.mean().plot(kind="bar")
```

```python
ab_dum=pd.get_dummies(df.copy())
mask = np.triu(ab_dum.corr())
ax = sns.heatmap(round(ab_dum.corr()*100,0)
                 ,annot=True
                 ,mask=mask
                 ,cmap='coolwarm')
ax.set_ylim((0,12))
ab_dum.shape
#plt.savefig('figures/correlogram.png')
```

```python
_ =sns.pairplot(ab_dum)
```
