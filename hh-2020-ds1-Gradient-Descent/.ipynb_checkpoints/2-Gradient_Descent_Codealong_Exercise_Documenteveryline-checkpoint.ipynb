Add some Gaussian noise to the data

```python
X = 2 * np.random.rand(100,1) # 100 random floats from standard curve with deviation 1 times 2
y = 4 +3 * X+np.random.randn(100,1) # calculate y values by multiplying X by 3 and adding 4 and adding a small jitter
                                    # by adding a value from the normal distribution for 100 values, rand n means it can be negative
    
    
```

#  Analytical way of Linear Regression

```python
X_b = np.c_[np.ones((100,1)),X] # new array of 1s and x values (1s for calculation fo constant)
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) # linear regression calculation of theta values
print(theta_best)
```

<h5>This is close to our real thetas 4 and 3. It cannot be accurate due to the noise I have introduced in data

```python
X_new = np.array([[0],[2]]) # x values of 0 and 2 to calculate during thetas and drawing the regression
X_new_b = np.c_[np.ones((2,1)),X_new] # concatenate (c_) new x values with an array of ones in front (for constant)
y_predict = X_new_b.dot(theta_best) # matrix product of x array with theta value to get the predicted y values
y_predict # print y values
```

<h5>Let's plot prediction line with calculated:theta

```python
plt.plot(X_new,y_predict,'r-') #plot red line (r-) with predicted 0 and predicted 2 values as start and end
plt.plot(X,y,'b.') # scatter plot of datra with blue dots (b.)
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18) # labels in latex stle
plt.axis([0,2,0,15]) # define coordinate system
```

# Gradient Descent

## Cost Function & Gradients

<h4> The equation for calculating cost function and gradients are as shown below. Please note the cost function is for Linear regression. For other algorithms the cost function will be different and the gradients would have  to be derived from the cost functions



<b>Cost</b>
\begin{equation}
J(\theta) = 1/2m \sum_{i=1}^{m} (h(\theta)^{(i)} - y^{(i)})^2 
\end{equation}

<b>Gradient</b>

\begin{equation}
\frac{\partial J(\theta)}{\partial \theta_j} = 1/m\sum_{i=1}^{m}(h(\theta^{(i)} - y^{(i)}).X_j^{(i)}
\end{equation}

<b>Gradients</b>
\begin{equation}
\theta_0: = \theta_0 -\alpha . (1/m .\sum_{i=1}^{m}(h(\theta^{(i)} - y^{(i)}).X_0^{(i)})
\end{equation}
\begin{equation}
\theta_1: = \theta_1 -\alpha . (1/m .\sum_{i=1}^{m}(h(\theta^{(i)} - y^{(i)}).X_1^{(i)})
\end{equation}
\begin{equation}
\theta_2: = \theta_2 -\alpha . (1/m .\sum_{i=1}^{m}(h(\theta^{(i)} - y^{(i)}).X_2^{(i)})
\end{equation}

\begin{equation}
\theta_j: = \theta_j -\alpha . (1/m .\sum_{i=1}^{m}(h(\theta^{(i)} - y^{(i)}).X_0^{(i)})
\end{equation}

```python
#define cost calculation
def  cal_cost(theta,X,y):
    '''
    
    Calculates the cost for given X and Y. The following shows and example of a single dimensional X
    theta = Vector of thetas 
    X     = Row of X's np.zeros((2,j))
    y     = Actual y's np.zeros((2,1))
    
    where:
        j is the no of features
    '''
    
    m = len(y) #number of values for calculation of mean
    
    predictions = X.dot(theta) #predict values with current theta, same as above
    cost = 1/(2*m) * np.sum(np.square(predictions-y)) #calculate mse and return as cost
    return cost
```

```python
# define gradient descent function
def gradient_descent(X,y,theta,learning_rate=0.01,iterations=100): #default values for learning rate and iterations
    '''
    X    = Matrix of X with added bias units
    y    = Vector of Y
    theta=Vector of thetas np.random.randn(j,1)
    learning_rate 
    iterations = no of iterations
    
    Returns the final theta vector and array of cost history over no of iterations
    '''
    m = len(y) 
    #number of observations
    cost_history = np.zeros(iterations) 
    # prepare numpy array for recording cost history
    theta_history = np.zeros((iterations,2)) 
    # prepare numpy array for recording both thetas (hence 2d array)
    for it in range(iterations): 
        # loop number of times. as iterations
        prediction = np.dot(X,theta)
        # take previous (or start) theta and predict using whole x array
        theta = theta -(1/m)*learning_rate*( X.T.dot((prediction - y))) 
        # adjust theta by gradient
        theta_history[it,:] =theta.T
        #record both thetas at index of current iteration, transpose to get them side by side
        cost_history[it]  = cal_cost(theta,X,y)
        # record mse in cost history at current iteration index
        
    return theta, cost_history, theta_history
    #return final theta, cost bistory and theta history
        
    
    
```

<h3> Let's start with 1000 iterations and a learning rate of 0.01. Start with theta from a Gaussian distribution

```python
theta = np.random.randn(2,1)
theta
```

```python
lr =0.01
#define learning rate
n_iter = 1000
# define n of iterations
theta = np.random.randn(2,1)
# random start theta
X_b = np.c_[np.ones((len(X),1)),X]
# prepare x array with ones in front
theta,cost_history,theta_history = gradient_descent(X_b,y,theta,lr,n_iter)
# execute gradient descent function with prepared values


print('Theta0:          {:0.3f},\nTheta1:          {:0.3f}'.format(theta[0][0],theta[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
# print out final thetas and final cost
```

<h3> Let's plot the cost history over iterations

```python
fig,ax = plt.subplots(figsize=(12,8))
# prepare plot
ax.set_ylabel('J(Theta)')
ax.set_xlabel('Iterations') #set labels
ax.plot(range(n_iter),cost_history,'b.');
#plot cost history
```

<h3> After around 150 iterations the cost is flat so the remaining iterations  are not needed or will not result in any further optimization. Let us zoom in till iteration 200 and see the curve

```python

fig,ax = plt.subplots(figsize=(10,8))
ax.plot(range(200),cost_history[:200],'b.');
# plot cost history till iteration 200
```

<b>It is worth while to note that the cost drops faster initially and then the gain in cost reduction is not as much

### It would be great to see the effect of different learning rates and iterations together

### Let us  build a function which can show the effects together and also show how gradient decent actually is working

```python
# define function for plotting scatter plot of data with lines for different thetas and optional plot for cost history
def plot_GD(n_iter,lr,ax,ax1=None):
     """
     n_iter = no of iterations
     lr = Learning Rate
     ax = Axis to plot the Gradient Descent
     ax1 = Axis to plot cost_history vs Iterations plot

     """
     ax.plot(X,y,'b.')
    #plot scatter of data   
     theta = np.random.randn(2,1)
    #set random start theta
     tr =0.1
     #start seed for transparency
     cost_history = np.zeros(n_iter)
    #prepare empty cost history
     for i in range(n_iter):
        #loop for number of iterations
        #pred_prev = X_b.dot(theta)
        # predict y values of x array with current theta
        theta,h,_ = gradient_descent(X_b,y,theta,lr,1)
        #update theta using gradient descent with 1 iteration using current theta
        pred = X_b.dot(theta)
        # predict using updated thheta
        cost_history[i] = h[0]
        # update cost history
        if ((i % 25 == 0) ):
        #every 25 steps
            ax.plot(X,pred,'r-',alpha=tr)
            # plot red line using current predictions with opacity tr
            if tr < 0.8:
                tr = tr+0.2
                #update opacity bei 0.2
     if not ax1== None:
        #if ax1 is defined
        ax1.plot(range(n_iter),cost_history,'b.') 
        #plot cost history
```

### Plot the graphs for different iterations and learning rates combination

```python
fig = plt.figure(figsize=(30,25),dpi=200)
#prepare figuresize and dpi
fig.subplots_adjust(hspace=0.4, wspace=0.4)
#prepare plotting of subplots
it_lr =[(2000,0.001),(500,0.01),(200,0.05),(100,0.1)]
# list of tuples where each tuple is number of iterations and learning rate
count =0
#prepare rolling index
for n_iter, lr in it_lr:
    #loop for each tuple
    count += 1
    #update rolling count
    ax = fig.add_subplot(4, 2, count)
    #plot subplot 
    count += 1
    #plot subplot
    ax1 = fig.add_subplot(4,2,count)
    # plot second subplot
    ax.set_title("lr:{}".format(lr))
    # first subplot title with learning rate
    ax1.set_title("Iterations:{}".format(n_iter))
    # second suplot title with iterations
    plot_GD(n_iter,lr,ax,ax1)
    #use plot gd to plot the plots for each tuple
    
```

<b> See how useful it is to visualize the effect of learning rates and iterations on gradient descent. The red lines show how the gradient descent starts and then slowly gets closer to the final value

# Stochastic Gradient Descent

```python
# define function for stochastic descent
def stocashtic_gradient_descent(X,y,theta,learning_rate=0.01,iterations=10):
    '''
    X    = Matrix of X with added bias units
    y    = Vector of Y
    theta=Vector of thetas np.random.randn(j,1)
    learning_rate 
    iterations = no of iterations
    
    Returns the final theta vector and array of cost history over no of iterations
    '''
    m = len(y)
    # number of obs
    cost_history = np.zeros(iterations)
    # prepare empty cost history
    
    for it in range(iterations):
        #loop times number of iteration
        cost =0.0
        # set cost to 0 at beginning of loop
        for i in range(m):
            # new loop for number of obs, in order to use cal cost function
            # since it divides by m
            rand_ind = np.random.randint(0,m)
            # generate random integer between 0 and m to randomly select observation from array
            X_i = X[rand_ind,:].reshape(1,X.shape[1])
            # use integer to select x value
            y_i = y[rand_ind].reshape(1,1)
            # use integer to select corresponding "real" y value
            prediction = np.dot(X_i,theta)
            #predict as above using selected x and current theta
            theta = theta -(1/m)*learning_rate*( X_i.T.dot((prediction - y_i)))
            #adjust theta by learning rate
            cost = cal_cost(theta,X_i,y_i)
            # calculate cost and make a rolling addition of that cost
            # since it is done m times by the loop we have to do a rolling addition
            # and accumulate because the cost of the single prediciton is each time divided
            # by m
        cost_history[it]  += cost
        #add accumulated cost to history
    return theta, cost_history
    # return theta and cost history
```

```python
lr =0.5
# predefine lr
n_iter = 100
# predefine iteration
theta = np.random.randn(2,1)
#random start theta
X_b = np.c_[np.ones((len(X),1)),X]
# prepare x array
theta,cost_history = stocashtic_gradient_descent(X_b,y,theta,lr,n_iter)
#calculate theta and cost history

print('Theta0:          {:0.3f},\nTheta1:          {:0.3f}'.format(theta[0][0],theta[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
```

```python
fig,ax = plt.subplots(figsize=(10,8))

ax.set_ylabel('{J(Theta)}',rotation=0)
ax.set_xlabel('{Iterations}')
theta = np.random.randn(2,1)
#plot cost history same as before
_=ax.plot(range(n_iter),cost_history,'b.')
```

# Mini Batch Gradient Descent

```python
np.random.permutation(100)
```

```python
def minibatch_gradient_descent(X,y,theta,learning_rate=0.01,iterations=10,batch_size =20):
    '''
    X    = Matrix of X without added bias units
    y    = Vector of Y
    theta=Vector of thetas np.random.randn(j,1)
    learning_rate 
    iterations = no of iterations
    
    Returns the final theta vector and array of cost history over no of iterations
    '''
    m = len(y)
    # number of obs
    cost_history = np.zeros(iterations)
    # prepare empty cost history
    n_batches = int(m/batch_size)
    # calculate number of batches by dividing obs by batch size
    for it in range(iterations):
        # loop number of times of iteration
        cost =0.0
        # set rolling cost to 0
        indices = np.random.permutation(m)
        # get numbers of 0 to n obs in random order
        X = X[indices]
        # update x values by reordering them according to random numbers generated
        y = y[indices]
        # same with y 
        for i in range(0,m,batch_size):
            #loop number of times of observation divided by batch size (eg 100 obs  batch size 20, do this 5 times)
            # at i 0 20 40 60 and 80
            X_i = X[i:i+batch_size]
            #create x batch at i to i + batch size, eg: 0:20, 20:40 ... if using above example
            y_i = y[i:i+batch_size]
            #same for y 
            X_i = np.c_[np.ones(len(X_i)),X_i]
            # prepare x batch array with ones for prediciton calc
            prediction = np.dot(X_i,theta)
            # predict y values for batch with current theta
            theta = theta -(1/m)*learning_rate*( X_i.T.dot((prediction - y_i)))
            #adjust theta
            cost_history[it] += cal_cost(theta,X_i,y_i)/n_batches
            # rolling cost history update
        #cost_history[it]  = cost
        
    return theta, cost_history
```

```python
lr =0.1
n_iter = 200

theta = np.random.randn(2,1)


theta,cost_history = minibatch_gradient_descent(X,y,theta,lr,n_iter)
#do gradient descent as above

print('Theta0:          {:0.3f},\nTheta1:          {:0.3f}'.format(theta[0][0],theta[1][0]))
print('Final cost/MSE:  {:0.3f}'.format(cost_history[-1]))
```

```python
fig,ax = plt.subplots(figsize=(10,8))

ax.set_ylabel('{J(Theta)}',rotation=0)
ax.set_xlabel('{Iterations}')
theta = np.random.randn(2,1)
# plot history as above
_=ax.plot(range(n_iter),cost_history,'b.')
```
