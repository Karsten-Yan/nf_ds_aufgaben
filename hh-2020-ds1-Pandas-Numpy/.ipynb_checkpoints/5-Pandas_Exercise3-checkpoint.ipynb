3) I've decided that this month I want to stay away from wines with relatively high alcohol content. To do that, I'm going to avoid any wines that have a greater alcohol content than the mean alcohol content, and you're going to help me do this. To achieve this, let's do the following: 

```python
import pandas as pd
dfred = pd.read_csv("data/winequality-red.csv", delimiter=";")
dfwhite = pd.read_csv("data/winequality-white.csv", delimiter =";")

cols = dfred.columns.to_list()
cols = [col.replace(" ","_") for col in cols]

dfred.columns = cols
dfwhite.columns = cols

dfred.head()
```

  * Find the mean alcohol content, separately, for reds and whites.  
  

```python
dfred.alcohol.mean()
```

```python
dfwhite.alcohol.mean()
```

  * Create a `Series` that holds whether each row in each `DataFrame` (the reds, and whites) has a higher alcohol content than the mean. 
  

```python
dfred_alc_series =dfred["alcohol"] >= dfred.alcohol.mean()
dfred_alc_series= dfred_alc_series.rename("alcohol_over_mean")
```

```python
dfwhite_alc_series =dfwhite["alcohol"] >= dfwhite.alcohol.mean()
dfwhite_alc_series = dfwhite_alc_series.rename("alcohol_over_mean")
```

  * Merge this `Series` onto the `DataFrame`. I can imagine doing this with either a `.join()` or using `pd.concat()`. For practice, do it with both. Note: merges with `Series` work the same way that they work with `DataFrames`.  
  

```python
dfred_joined = dfred.join(dfred_alc_series)
dfred_joined.head()
```

```python
dfred_joined2= pd.concat([dfred, dfred_alc_series], axis=1)
dfred_joined2.head()
```

```python
dfwhite_joined = dfwhite.join(dfwhite_alc_series)
dfwhite_joined.head()
```

```python
dfwhite_joined2= pd.concat([dfwhite, dfwhite_alc_series], axis=1)
dfwhite_joined2.head()
```

  * Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.  
   

```python
red_low = dfred_joined[dfred_joined["alcohol_over_mean"] == False]

red_low.head()
```

```python
white_low = dfwhite_joined[dfwhite_joined["alcohol_over_mean"] == False]
white_low.head()
```

   
4) Let's say that I want to get started on cutting back next month. This time, though, I want to focus on staying away from those wines with a high acidity. Specifically, I want to stay away from those wines that are in the highest bin of fixed acidity (highest bin out of 5). You're now going to help me do this. To achieve this, let's do the following: 

 * Separate the rows in each `DataFrame` into 5 equal width bins based off their fixed acidity. 
 

```python
red_acid_bins = pd.cut(dfred_joined.fixed_acidity, 5)
red_acid_bins = red_acid_bins.rename("fixed_acidity_bins")
red_acid_bins.head()
```

```python
white_acid_bins = pd.cut(dfwhite_joined.fixed_acidity, 5)
white_acid_bins = white_acid_bins.rename("fixed_acidity_bins")
white_acid_bins.head()
```

 * Merge the resulting `Series` holding these 5 bins onto the original `DataFrame`. I can imagine also doing this with either `.join()` or using `pd.concat()`. Try doing it with both for practice. 

```python
dfred_joined_3 = dfred_joined.join(red_acid_bins)
dfred_joined_3.fixed_acidity_bins.unique()
dfred_joined_3.fixed_acidity_bins.max()
```

```python
dfwhite_joined_3 = dfwhite_joined.join(white_acid_bins)
dfwhite_joined_3.fixed_acidity_bins.unique()
```

 * Return back to me all those rows that are **not in ** the top bin in terms of fixed acidity. 
 

```python
dfred_joined_3[dfred_joined_3.fixed_acidity_bins != dfred_joined_3.fixed_acidity_bins.max()]
```

```python
dfwhite_joined_3[dfwhite_joined_3.fixed_acidity_bins != dfwhite_joined_3.fixed_acidity_bins.max()]
```

```python
dfred_joined_3.groupby("fixed_acidity_bins").count()
```

 * The average `alcohol` content for those reds above the mean `alcohol` level, by quality.
 * The average `alcohol` content for those whites above the mean `alcohol` level, by quality. 
 
 Use a `pivot table` to solve this. 
 
 

```python
pd.pivot_table(dfred_joined_3[dfred_joined_3["alcohol_over_mean"] == True], index = dfred_joined_3.quality)
```

```python
 pd.pivot_table(dfred_joined_3[dfred_joined_3["alcohol_over_mean"] == True][["alcohol"]], index = dfred_joined_3.quality)
```

```python
pd.pivot_table(dfwhite_joined_3[dfwhite_joined_3["alcohol_over_mean"] == True][["alcohol"]], index = dfwhite_joined_3.quality)
```
